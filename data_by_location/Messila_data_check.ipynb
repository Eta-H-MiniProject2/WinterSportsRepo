```python
import pandas as pd
import numpy as np

messila_df = pd.read_csv("Messila_data.csv")
messila_df
```

```python
messila_df.isna().sum()
```

```python
cloud_coverage_nan = messila_df[messila_df['cloud_cover_rate'].isna()]
cloud_coverage_nan
```

removing duplicates as cloud coverage data from 2 stations. Examples why needs to be removed:
- 2025-03-10,6,-,0.4,,-
- 2025-03-10,6,-,0.4,8,Cloudy

```python

messila_df = messila_df.drop_duplicates(subset=['date'], keep='last')
messila_df
```

```python
#replacing all "-"" values with NaN
messila_df = messila_df.replace('-', np.nan)
messila_df
```

```python
cloud_coverage_nan = messila_df[messila_df['cloud_cover_rate'].isna()]
cloud_coverage_nan
```

```python
# filling a NaN value in avg_temp_c to same as the day before
messila_df['avg_temp_c'] = messila_df['avg_temp_c'].fillna('2.2')
messila_df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

target = messila_df['date']
feature_list = ["avg_temp_c"]

x_train, x_test, y_train, y_test = train_test_split(messila_df[feature_list], target, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

predict = model.predict(x_test)

loss = np.sqrt(mean_squared_error(y_test, predict))
print('Loss:', loss)
```
